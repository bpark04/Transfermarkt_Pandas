## Data Preprocessing

```python
# deleting unnecessary characters in value column and change its datatype to int
# €100.00m -> 100.00
df['value'] = df['value'].str.replace("€",'')
df['value'] = df['value'].str.replace("m",'').astype('float')
df.head()
```

```python
# new column
df['in cad(m)'] = df['value']*1.48
df.head()
```
